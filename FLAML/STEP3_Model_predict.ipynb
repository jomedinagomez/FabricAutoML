```python
import mlflow.pyfunc

df = spark.read.format("delta").load(
    "Tables/class_transformed_taxi_data"
)

model_name = "classification_demo"
model_version = 'latest'
output_col = 'class_predicted'

input_columns = ["vendor","store_forward","pickup_longitude","pickup_latitude","dropoff_longitude","dropoff_latitude","passengers","distance","pickup_weekday","pickup_month","pickup_monthday","pickup_hour","pickup_minute","pickup_second","dropoff_weekday","dropoff_month","dropoff_monthday","dropoff_hour","dropoff_minute","dropoff_second"]

model = mlflow.pyfunc.load_model(model_uri=f"models:/{model_name}/{model_version}")

predictions = model.predict(df.toPandas()[input_columns])

if hasattr(predictions[0], "__len__"):
    print('1')
    new_pred = [int(single_output[0]) for single_output in predictions]
else:
    new_pred = predictions

df = df.toPandas()

df[output_col] = new_pred
df=spark.createDataFrame(df) 

df.write.format('delta').mode("overwrite").save(
    "Tables/table_model_output"
)
```
